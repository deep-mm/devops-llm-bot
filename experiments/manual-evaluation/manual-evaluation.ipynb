```python
# Read three CSV files from ../final-results/gpt-4-1.csv, ../final-results/gpt-4-2.csv, ../final-results/gpt-4-3.csv of format Index,GitHub_Repo_Link,GitHub_Build_Pipeline_File_Content,Generated_Build_Pipeline_File_Content,Exact_Match_Score,BLEU_Score,Syntax_Check,DevOps_Aware_Score,Language
# Randomly select 324 rows of all the rows with Syntax_Check as Valid from combined CSV files and save 108 rows each into three new CSV files in ../final-results/manual-validation with the names manual-validation-1.csv, manual-validation-2.csv, manual-validation-3.csv
# In each csv file add a new column named "Manual_Validation_Score"
```

```python
import pandas as pd
import random
import os 

# Read the CSV files
csv_files = ['../final-results/gpt-4-1.csv', '../final-results/gpt-4-2.csv', '../final-results/gpt-4-3.csv']
dfs = [pd.read_csv(file) for file in csv_files]
```

```python
valid_count = pd.concat(dfs)['Syntax_Check'].value_counts()['Valid']
print(f"Count of valid values in 'Syntax_Check' column: {valid_count}")
```

```python
# Filter rows with Syntax_Check as Valid
valid_rows = pd.concat(dfs).loc[pd.concat(dfs)['Syntax_Check'] == 'Valid']

# Randomly select 324 rows
random_rows = valid_rows.sample(n=324, random_state=42)
```

```python
random_rows 
```

```python
# Split the rows into three groups
groups = [random_rows.iloc[i:i+108] for i in range(0, len(random_rows), 108)]

# Save the groups into new CSV files
output_files = ['./manual-evaluation-1.csv',
                './manual-evaluation-2.csv',
                './manual-evaluation-3.csv']

for i, group in enumerate(groups):
    group['Manual_Evaluation_Score'] = ''
    group.to_csv(output_files[i], index=False)
```

```python
selected_repositories = random_rows.copy()
selected_repositories.reset_index(drop=True, inplace=True)
selected_repositories['Index'] = selected_repositories.index
selected_repositories['Manual_Evaluation_Score'] = 0
```

```python
selected_repositories
```

```python
# Create a folder named 'manual_eval_dataset'
dataset_folder = 'manual_eval_dataset'
os.makedirs(dataset_folder, exist_ok=True)

# Create a folder for each repository inside the 'index' folder
for index, row in selected_repositories.iterrows():
    folder_name = os.path.join(dataset_folder, str(row['Index']))
    os.makedirs(folder_name, exist_ok=True)
    
    # Create ground_truth.yaml file
    ground_truth_file = os.path.join(folder_name, 'ground_truth.yaml')
    with open(ground_truth_file, 'w') as file:
        file.write(row['GitHub_Build_Pipeline_File_Content'])
    
    # Create generated.yaml file
    generated_file = os.path.join(folder_name, 'generated.yaml')
    with open(generated_file, 'w') as file:
        file.write(row['Generated_GitHub_Build_Pipeline_File_Content'])
    
    # Create text file with repository link and index column value
    text_file = os.path.join(folder_name, 'info.txt')
    with open(text_file, 'w') as file:
        file.write(f"Repository Link: {row['GitHub_Repo_Link']}\n")
        file.write(f"Index: {row['Index']}\n")
```

```python
# Create the "manual_eval_dataset" folder if it doesn't exist
dataset_folder = 'manual_eval_dataset'
os.makedirs(dataset_folder, exist_ok=True)

# Define the file path
csv_file_path = os.path.join(dataset_folder, 'selected_repositories.csv')

# Save the DataFrame to the CSV file
selected_repositories.to_csv(csv_file_path, index=False)
```
